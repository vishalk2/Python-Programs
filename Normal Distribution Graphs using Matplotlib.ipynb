# Plotting 3 Normal Distributions on the same graph using Matplotlib

```python
#Import required Libraries(NumPy, Matplotlib, SciPy) to begin with.
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import math
```

```python
#Values of mean and variance for Normal Distribution are as follows:

#Set-1
mu_1 = 0
variance_1 = 5
sigma_1 = math.sqrt(variance_1)

#Set-2
mu_2 = 4
variance_2 = 0.2
sigma_2 = math.sqrt(variance_2)

#Set-3
mu_3 = -3.5
variance_3 = 0.05
sigma_3 = math.sqrt(variance_3)
```

```python
#X-variables: x = np.linspace(mean-3*sigma, mean+3*sigma, num=100)
x_1 = np.linspace(mu_1-3*sigma_1,mu_1+3*sigma_1,100)

x_2 = np.linspace(mu_2-3*sigma_2,mu_2+3*sigma_2,100)

x_3 = np.linspace(mu_3-3*sigma_3,mu_3+3*sigma_3,100)


#Y-variables: y = norm.pdf(x,mu,sigma)
y_1 = norm.pdf(x_1,mu_1,sigma_1)

y_2 = norm.pdf(x_2,mu_2,sigma_2)

y_3 = norm.pdf(x_3,mu_3,sigma_3)
```

```python
#Plotting the Normal Distribution graph (3-in-1)

#Labeling the axes
plt.xlabel('Value')
plt.ylabel('Normal Distribution')

#Labeling and plotting the curves
plt.plot(x_1,y_1,label='mu = 0, variance = 5')
plt.plot(x_2,y_2,label='mu = 4, variance = 0.2')
plt.plot(x_3,y_3,label='mu = -3.5, variance = 0.05')

#Adding a legend to the graph
plt.legend(loc='upper right')
```
