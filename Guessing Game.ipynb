```python
import random
number = random.randint(1,100)
```

```python
def guessing_game(number):
    
    #Printing out Intro & Rules of the Game
    print("All Aboard to the Guessing Game!")
    print('The rules to this game are basically as follows:')
    print("I'm thinking of a number between 1 and 100. All your task is to guess it in as little guesses as possible!")
    print('Here are some hints I will put out as you play:')
    print("If your guess is more than 10 away from my number, I'll tell you that you are COLD!")
    print("If your guess is within 10 of my number, I'll tell you that you are WARM!")
    print("If your guess goes farther than your most recent guess, I'll say you're getting COLDER!")
    print("If your guess goes closer than your most recent guess, I'll say you're getting WARMER!")
    print("Get it? WARM & WARMER as you near my guess; COLD & COLDER as you move away.")
    print("SO, here we go!")
    
    #Initializing with a Guess List
    guesses = [0]
    
    while True:
        
        #Taking input of from the User
        guess = int(input("I'm thinking of a number between 1 and 100.\n  What is your guess? "))
        
        #Checking if the input number is within the range of (1,100) or out of range
        if guess < 1 or guess > 100:
            print('OUT OF RANGE! Please try again within (1,100): ')
            continue

        # Comparing the player's guess to our number
        if guess == number:
            print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!!')
            break

        # If a guess is incorrect, adding it to the guesses list
        guesses.append(guess)

        # when testing the first guess, guesses[-2]==0, which evaluates to False
        # and brings us down to the second section
        
        # Checks if the current guess is closer/farther to the number 
        # as compared to the recent guess/previous guess
        if guesses[-2]:  
            if abs(number-guess) < abs(number-guesses[-2]):
                print('WARMER!')
            else:
                print('COLDER!')
        
        # Checks if the current guess is within a range of 10 with the number
        else:
            if abs(number-guess) <= 10:
                print('WARM!')
            else:
                print('COLD!')
```

```python
guessing_game(number)
```
