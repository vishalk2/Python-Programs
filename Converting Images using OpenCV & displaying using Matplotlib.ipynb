# Converting Images from BGR to RGB, HSC,HSL, inverts using OpenCV & displaying using Matplotlib

```python
#Importing both OpenCV & Pillow Libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#Defining a function so as to ease the process of calling it again if needed further
def image_conversions():
    #capturing the image through Webcam
    cap = cv2.VideoCapture(0)
    
    if(cap.isOpened()):
        ret,frame = cap.read()
        #print(ret)
        #print(frame)
    else:
        ret = False
    
    #1st Pic: BGR_to_RGB converted Image
    img1 = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) #gives a 3-Dimensional array
    plt.imshow(img1)
    plt.title('Color image in RGB')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    #can use cv2.imshow(window_name,image) if required to display using OpenCV
    
    #2nd Pic: RGB_to_BGR converted Image
    img2 = cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)
    plt.imshow(img2)
    plt.title('Color image in BGR')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    #3rd Pic: BGR_to_HSV converted Image
    img3 = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    plt.imshow(img3)
    plt.title('Color image in HSV')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    #4th Pic: BGR_to_HLS converted Image
    img4 = cv2.cvtColor(frame,cv2.COLOR_BGR2HLS)
    plt.imshow(img4)
    plt.title('Color image in HLS')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    #5th Pic: Inverted image
    img5 = cv2.bitwise_not(img1)
    plt.imshow(img5)
    plt.title('Inverted image')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    cap.release()
```

```python
image_conversions()
```
