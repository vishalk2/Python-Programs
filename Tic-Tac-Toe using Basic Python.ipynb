# Tic-Tac-Toe

```python
from IPython.display import clear_output

#Setting up the board

def board(board_inputs):
    clear_output()
    
    print('-------------')
    print('|   |   |   |')
    print('| ' + board_inputs[7] + ' | ' + board_inputs[8] + ' | ' + board_inputs[9] + ' |')
    print('|   |   |   |')
    print('-------------')
    print('|   |   |   |')
    print('| ' + board_inputs[4] + ' | ' + board_inputs[5] + ' | ' + board_inputs[6] + ' |')
    print('|   |   |   |')
    print('-------------')
    print('|   |   |   |')
    print('| ' + board_inputs[1] + ' | ' + board_inputs[2] + ' | ' + board_inputs[3] + ' |')
    print('|   |   |   |')
    print('-------------')
    
    #The board will be displayed in the following format:
    #-------------
    #|   |   |   |
    #| X | O | X |
    #|   |   |   |
    #-------------
    #|   |   |   |
    #| O | X | O |
    #|   |   |   |
    #-------------
    #|   |   |   |
    #| X | O | X |
    #|   |   |   |
    #-------------
    
board_inputs = ['#','X','O','X','O','X','O','X','O','X']
#board(board_inputs)
```

```python
#Asking player for Input

def player_input():
    marker = ''
    while not (marker == 'X' or marker == 'O'):
        marker = input('Choose one out of "X" or "O" for the game and enter your input here without the quotes> ').upper()
    if(marker == 'X'):
        print('Player_1 = "X" & Player_2 = "O"')
        return ('X','O')
    else:
        print('Player_1 = "O" & Player_2 = "X"')
        return ('O','X')

#player_input()
```

```python
#Assigning a position to the marker on the board

def place_of_marker(board_inputs,marker,position):
    board_inputs[position]=marker
    
#place_of_marker(board_inputs,'$',8)
#board(board_inputs)
```

```python
def win_checl(board,mark):
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or
      (board[4] == mark and board[5] == mark and board[6] == mark) or
      (board[1] == mark and board[2] == mark and board[3] == mark) or
      (board[7] == mark and board[4] == mark and board[1] == mark) or
      (board[8] == mark and board[5] == mark and board[2] == mark) or
      (board[9] == mark and board[6] == mark and board[3] == mark) or
      (board[7] == mark and board[5] == mark and board[3] == mark) or
      (board[9] == mark and board[5] == mark and board[1] == mark))

#win_check(board_inputs,'X')
```

```python
#Deciding which player to go first with a random generator
import random

def first_player():
    a=random.randint(0,1)
    if(a == 0):
        return 'Player 2'
    else:
        return 'Player 1'

#first_player()
```

```python
#Determing whether a space/empty slot is available on the board
def space_check(board,position):
    return board[position] == ' '

#space_check(board_inputs,4)
```

```python
#Determing if the board is fully occupied or not
def board_check(board):
    for i in range(1,10):
        if(space_check(board,i)):
            return False
    return True
#board_check(board_inputs)
```

```python
#Asking the player for the next move
def player_chance(board):
    position=0
    while position not in [i for i in range(1,10)] or not space_check(board,position):
        position = int(input('Enter next position on the board: (1-9)'))
    return position
```

```python
#Asking for a replay of the game
def replay():
    return input('Want to play another game? Type "Yes" if willing else type "No" >').lower().startswith('y')
```

# The GAME!

```python
print('Hello! Welcome to Tic-Tac-Toe with Python')

while True:
    the_board = [' ']*10
    player_1,player_2 = player_input()
    
    whose_turn = first_player()
    print(whose_turn + ' will go first! Rock & Roll! ;)')
    
    play_game = input('Type "Yes" to proceed to the game else "No" without quotes >').lower()
    if(play_game == 'yes'):
        game_on = True
    else:
        game_on = False
        break
    
    while game_on:
        if(whose_turn == 'Player 1'):
            board(the_board)
            position = player_chance(the_board)
            place_of_marker(the_board,player_1,position)
            
            if(win_check(the_board,player_1)):
                board(the_board)
                print('Hurray! The game is won by {}'.format('Player_1'))
                game_on = False
            else:
                if(board_check(the_board)):
                    board(the_board)
                    print('The Game is a Draw! :(')
                    break
                else:
                    whose_turn = 'Player 2'
        else:
            board(the_board)
            position = player_chance(the_board)
            place_of_marker(the_board,player_2,position)
            
            if(win_check(the_board,player_2)):
                board(the_board)
                print('Congrats {} ! You have won!'.format('Player_2'))
                game_on = False
            else:
                if(board_check(the_board)):
                    board(the_board)
                    print('The Game is a Draw! :(')
                    break
                else:
                    whose_turn = 'Player 1'
    if not replay():
        break
```
